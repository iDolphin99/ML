```python
X_column = ['Pclass', 'Sex', 'SibSp', 'Parch',  'Embarked', 'Survived']
y_column = 'Age'  # 정답값, y, true value, target value, label

# 모델에 사용할 컬럼(column, feature) 결정 
input_data = df[ X_column + [y_column] ].copy()
input_data
```

```python
input_data["Cat_Age"] = "0"
input_data.loc[(input_data["Age"] >= 10 ) & (input_data["Age"] < 20), "Cat_Age"] = "10"
input_data.loc[(input_data["Age"] >= 20 ) & (input_data["Age"] < 30), "Cat_Age"] = "20"
input_data.loc[(input_data["Age"] >= 30 ) & (input_data["Age"] < 40), "Cat_Age"] = "30"
input_data.loc[(input_data["Age"] >= 40 ) & (input_data["Age"] < 50), "Cat_Age"] = "40"
input_data.loc[(input_data["Age"] >= 50 ) , "Cat_Age"] = "50"
```

```python
# categorical columns extraction
#cat_cols = ['Sex', 'Embarked']
cat_cols = input_data.columns[input_data.dtypes == object]
cat_cols
```

```python
# label encoding (integer encoding)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

input_data["Cat_Age_Str"] = input_data["Cat_Age"]

for c in cat_cols:
    input_data[c] = le.fit_transform(list(input_data[c]))  # list() 로 씌우면 결측치 자동 처리
```

```python
# Embarked
print(le.classes_)
print(input_data['Cat_Age'].unique())
```

```python
X_column = ['Pclass', 'Sex', 'SibSp', 'Parch',  'Embarked', 'Survived']
y_column = 'Cat_Age'  # 정답값, y, true value, target value, label
```

```python
# 결측치 체크
input_data.columns[input_data.isna().any()].tolist()
```

```python
# 결측치 처리
# input_data = input_data.fillna(-1) 
```

```python
# 데이터 -> train/valid  ,  test
from sklearn.model_selection import train_test_split

tr_val_X, test_X, tr_val_y, test_y = train_test_split(
    input_data[X_column],   # X 입력데이터
    input_data[y_column],   # y 정답데이터
    test_size=0.2,
    shuffle=True,
    random_state=42,
    stratify=input_data[y_column]   # classification 일때는 중요한 옵션
    )
```

```python
tr_val_data = pd.concat([tr_val_X, tr_val_y], axis=1)
tr_val_data
```

```python
# 데이터 -> train/valid  ,  test
from sklearn.model_selection import train_test_split

train_X, valid_X, train_y, valid_y = train_test_split(
    tr_val_data[X_column],   # X 입력데이터
    tr_val_data[y_column],   # y 정답데이터
    test_size=0.2,
    #train_size=13 # 데이터 개수
    shuffle=True,
    random_state=42,
    stratify=tr_val_data[y_column]   # classification 일때는 중요한 옵션
    )
```

```python
print(len(train_X))
print(len(valid_X))
print(len(test_X))
len(train_X) + len(valid_X) + len(test_X) # 20
```

```python
from catboost import CatBoostClassifier
cb = CatBoostClassifier(verbose=100)
cb.fit(train_X, train_y, eval_set=[(valid_X, valid_y)])
```

```python
# classification prediction (예측)
pred_y_prob = cb.predict_proba(test_X)
pred_y_prob
```

```python
pred_y = np.argmax(pred_y_prob, axis=1)
pred_y
```

```python
test_y.values
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(test_y, pred_y))
```
